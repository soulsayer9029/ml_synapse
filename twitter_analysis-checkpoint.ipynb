```python
#removing punctuation marks
x['cleaned_tweets'] = x['cleaned_tweets'].str.replace("[^a-zA-Z#]", " ")
x.head(20)
```

```python
#Initialisibg lemmatizer and stopwords
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
 
lemmatizer=WordNetLemmatizer()
tweets=x['cleaned_tweets']
corpus=[]
tweets.head()
```

```python
#applied stopwords and lemmatization
for i in range(len(tweets)):
    tweet_cleaned=re.sub('[^a-zA-Z ]','',tweets[i])
    tweet_cleaned=tweet_cleaned.lower()
    tweet_cleaned=tweet_cleaned.split()
    tweet_cleaned = [lemmatizer.lemmatize(word) for word in tweet_cleaned if word not in set(stopwords.words('english'))]
    tweet_cleaned=' '.join(tweet_cleaned)
    corpus.append(tweet_cleaned)
print(corpus[3])
```

```python
#vectorizing data using TFIDF vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer(max_features=5000)  
X=cv.fit_transform(corpus).toarray()
```

```python
#creating train and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
#creating naive bayes model
from sklearn.naive_bayes import MultinomialNB
sentiment_detect_model=MultinomialNB()
sentiment_detect_model.fit(x_train,y_train)#fitting data to model
```

```python
#finding accuracy
y_pred=sentiment_detect_model.predict(x_test)

acc = sentiment_detect_model.score(x_test, y_test)
print(acc)

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#checking for overfitting
y_pred2=sentiment_detect_model.predict(x_train)
cm2 = confusion_matrix(y_train, y_pred2)
print(cm2)

```
